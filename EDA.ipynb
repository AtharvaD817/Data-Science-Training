```python
af['Installs'] = af['Installs'].str.split('+').str[0]
```

```python
af['Installs'].unique()
```

```python
af['Installs'].replace(',','', regex=True, inplace=True) 
```

```python
af['Installs'].unique()
```

```python
af['Installs'] = af['Installs'].astype(np.int64)
```

```python
af['Installs'].unique()
```

```python
af['Currency'].unique()

#remove the currency xxx
```

```python
af['Currency'] = af['Currency'].apply(lambda x: str(x).replace('XXX', 'USD') if 'XXX'in str(x) else x)
```

```python
af['Currency'].unique()
```

```python
af['Size'].unique()
```

```python
af['Size'] = af['Size'].apply(lambda x: str(x).replace('M', '')if 'M' in str(x) else x)
```

```python
af['Size']
```

```python
af['Size'] = af['Size'].apply(lambda x: str(x).replace(',', '.') if ',' in str(x) else x)
```

```python
af['Size'] = af['Size'].apply(lambda x: float(str(x).replace('k', '')) /1000 if 'k' in str(x) else x)
```

```python
af['Size']
```

```python
af['Size'] = af['Size'].apply(lambda x: str(x).replace('Varies with device', '0') if 'Varies with device' in str(x) else x)
```

```python
af['Size'] = af['Size'].apply(lambda x: float(str(x).replace('G', '')) * 1000 if 'G' in str(x) else x)
```

```python
af['Size'] = af['Size'].apply(lambda x: float(x))
```

```python
af.dtypes['Size']
```

```python
af['Free']
```

```python
af['Type'] = np.where(af['Free'] == True, 'Free','Paid')
af.drop(['Free'],axis=1, inplace= True )
```

```python
af.Type.unique()
```

```python
af['Content Rating'].unique()
```

```python
af['Content Rating'].value_counts()
```

```python
af['Content Rating'] = af['Content Rating'].replace('Unrated', 'Everyone')
af['Content Rating'] = af['Content Rating'].replace('Everyone 10+', 'Teen')
af['Content Rating'] = af['Content Rating'].replace('Mature 17+', 'Ault')
af['Content Rating'] = af['Content Rating'].replace('Everyone 10+', 'Ault')
```
